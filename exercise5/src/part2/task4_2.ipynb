### TODO: Generate data for Lorenz attractor (using prev exercise)

```python
# TODO: Generate lorenz_dataectory data by simulating the Lorenz attractor
initial = [10, 10, 10]
trajectory = lorenz_traj(initial,t=np.linspace(0, 1000, 100000))
print(trajectory.shape)
# Plotting the trajectory

# TODO: Visualize lorenz attractor
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(trajectory[:, 0], trajectory[:, 1], trajectory[:, 2], lw=0.25)

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('Lorenz Attractor Trajectory')
plt.show()
```

### TODO: Create and visualize time-delayed x coordinates from Lorenz data in a 3-d Euclidean space


```python
# Hint: Use method time_delay from utils.py
delay_coords_x = time_delay(data=trajectory, col=0, delta_t=10, out_dim=3)
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111,projection='3d')
ax.plot(delay_coords_x[:,0], delay_coords_x[:,1],delay_coords_x[:,2], lw=0.25)
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05)
#plt.tight_layout()
#ax.view_init(elev=20, azim=120)
ax.set_xlabel("X(n)")
ax.set_ylabel("X(n+10)")
ax.set_zlabel("X(n+20)")
plt.title("Time delay of coordinate x")
plt.show()
```
