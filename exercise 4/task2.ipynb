### Task 2

In this part, the bifurcation of the dynamical system $\dot{x} = \alpha - x^2$ is analyzed.

```python
np.seterr(over='ignore')

system = Task2(beta=1, gamma=0)
X = system.fun()
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()
fig, ax = plot_bifurcation_diagram_task_2(fig, ax, X,gamma=0, title="x' = α - x² Bifurcation Diagram")
```
