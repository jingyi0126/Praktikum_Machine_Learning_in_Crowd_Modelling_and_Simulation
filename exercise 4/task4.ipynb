## Part 2 Lorenz attractor

```python
# Parameters
sigma = 10
beta = 8/3
rho = 28
pars = [sigma, rho, beta]

# Domain
x_min = -50
x_max = 50
n_x = 21
y_min = -50
y_max = 50
n_y = 21
z_min = 0
z_max = 50
n_z = 11
list_par = [
    [x_min, x_max, n_x],
    [y_min, y_max, n_y],
    [z_min, z_max, n_z]
]

# System initialization
lorenz_attractor_high_rho = LorenzAttractor(pars, list_par)

# Creation of a trajectory
x_0 = np.array([10, 10, 10])
t_eval = np.linspace(0, 1000, 100001)

trajectory_high_rho = lorenz_attractor_high_rho.solve_system(
    fun=lorenz_attractor_high_rho.fun,
    init_state=x_0,
    t_eval=t_eval
)

# Plot the trajectory
plot_trajectory_lorenz_attractor(trajectory_high_rho.T, pars, noisy=False)
```

```python
# Creation of a trajectory for a noisy inital state
x_0_noisy = np.array([10 + 1e-8, 10, 10])

trajectory_high_rho_noisy = lorenz_attractor_high_rho.solve_system(
    fun=lorenz_attractor_high_rho.fun,
    init_state=x_0_noisy,
    t_eval=t_eval
)

# Plot the trajectory
plot_trajectory_lorenz_attractor(trajectory_high_rho_noisy.T, pars, noisy = True)
```

```python
# Plot the difference caused by noise
plot_error_lorenz_attractor(trajectory_high_rho, trajectory_high_rho_noisy, t_eval)
```

```python
# Change of rho
rho = 0.5
pars = [sigma, rho, beta]

# System initialization
lorenz_attractor_low_rho = LorenzAttractor(pars, list_par)

# Creation of a trajectory
x_0 = np.array([10, 10, 10])

trajectory_low_rho = lorenz_attractor_low_rho.solve_system(
    fun=lorenz_attractor_low_rho.fun,
    init_state=x_0,
    t_eval=t_eval
)

# Plot the trajectory
plot_trajectory_lorenz_attractor(trajectory_low_rho.T, pars, noisy=False)
```

```python
# Creation of a trajectory for a noisy inital state
x_0_noisy = np.array([10 + 1e-8, 10, 10])

trajectory_low_rho_noisy = lorenz_attractor_low_rho.solve_system(
    fun=lorenz_attractor_low_rho.fun,
    init_state=x_0_noisy,
    t_eval=t_eval
)

# Plot the trajectory
plot_trajectory_lorenz_attractor(trajectory_low_rho_noisy.T, pars, noisy = True)
```

```python
# Plot the difference caused by noise
plot_error_lorenz_attractor(trajectory_low_rho, trajectory_low_rho_noisy, t_eval)
```
