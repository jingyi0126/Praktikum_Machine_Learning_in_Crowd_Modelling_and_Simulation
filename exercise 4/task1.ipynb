### Task 1

### Parametrized matrix
For this task, we use the parametrized matrix $$A_{\alpha} = \begin{bmatrix} \alpha & \beta \\ |\beta| & \alpha \end{bmatrix}$$ with the eigenvalues $\lambda_{1/2} = \alpha \pm \sqrt{\beta \cdot |\beta|}$. Further, we chose $\beta = \alpha^3 - 3 \alpha$, which allows for the creation of all types of phase portraits.

### Stable node
The eigenvalues have to be real negative values. This can be achieved by choosing $0 < \beta < -\alpha$, i.e. $\alpha < 0$. To fulfil this requirement a value $\alpha \in (-\sqrt{3}, -\sqrt{2})$ has to be chosen.

```python
# # define parameters for the system
# evolution operator for the linear system
alpha = -1.65
beta = alpha ** 3 - 3 * alpha
A = np.array([
    [alpha, beta],
    [abs(beta), alpha]
])

# system domain
x_min = -1
x_max = 1
n_x = 100
y_min = -1
y_max = 1
n_y = 100
list_par = [
    [x_min, x_max, n_x],
    [y_min, y_max, n_y]
]

# # define the system
system = Task1(par=A, list_par_x=list_par)
X = system.X  # coordinate system
DX = system.fun(0, X)  # t is arbitrary here, since this system is time-independent

# # plot the phase portrait
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()
fig, ax = plot_phase_portrait_task_1(fig, ax, X, DX, "stable node", A)

# # plot a trajectory
# define parameters for solving the system
init_state = np.array([0, 1])  # init_state is an arbitrary state that leads to an interesting trajectory
t_eval = np.linspace(0, 100, 1000)

# solve the system
trajectory = system.solve_system(
    fun=system.fun,
    init_state=init_state,
    t_eval=t_eval
)

# plot trajectory
_ = ax.plot(trajectory[:, 0], trajectory[:, 1])
```
